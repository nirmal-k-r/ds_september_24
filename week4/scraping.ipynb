```python
from bs4 import BeautifulSoup as bs
```

```python
news_page_url='https://defimedia.info/categorie/actualites?page='
```

```python
page_num=0

final_url=news_page_url+str(page_num)
# print(final_url)
```

```python
#load the page
import requests
page=requests.get(final_url) #fetch the web page fromt he internet

# print(page)

# extract   
webPage=bs(page.content, 'html.parser')

print(webPage.prettify())
```

```python
article_directory=[]


#extract the section first
section=webPage.find('section', class_='col-sm-9 content-area')

base_url='https://defimedia.info'

# get all articles from that page section
articles=section.find_all('span', class_='sup-title')
for article in articles:
    title=article.text #get the text of the span tag
    url=article.parent['href'] #get the parent of the span tag and get the href attribute
    full_url=base_url+url
    article_directory.append({'title':title, 'url':full_url})

    
for article in article_directory:
    print(f'Article Title: {article["title"]} | Article URL: {article["url"]}')
   
```

```python
pages=10
article_directory=[]

for page_num in range(pages):
    final_url=news_page_url+str(page_num)

    page=requests.get(final_url)
    webPage=bs(page.content, 'html.parser')
    section=webPage.find('section', class_='col-sm-9 content-area')
    articles=section.find_all('span', class_='sup-title')

    for article in articles:
        title=article.text #get the text of the span tag
        url=article.parent['href'] #get the parent of the span tag and get the href attribute
        full_url=base_url+url
        article_directory.append({'title':title, 'url':full_url})
    
    print(f'Page {page_num} extracted. Length of article directory: {len(article_directory)}')


```

```python
#create a dataframe from the article directory
import pandas as pd
df=pd.DataFrame(article_directory)

df.head()
```

```python
df.to_csv('defimedia_articles.csv', index=False) #save the dataframe to a csv file
```
