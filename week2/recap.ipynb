```python
df2=df[['TypeName', 'Inches', 'ScreenResolution',
       'CPU_Company', 'CPU_Type', 'CPU_Frequency (GHz)', 'RAM (GB)', 'Memory',
       'GPU_Company', 'GPU_Type', 'OpSys', 'Weight (kg)', 'Price (Euro)']]

df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
columns_to_be_encoded=['TypeName', 'ScreenResolution','CPU_Company', 'CPU_Type', 'Memory','GPU_Company', 'GPU_Type', 'OpSys']

# #encode columns
# for column in columns_to_be_encoded:
#     df2[column]=df2[column].astype('category').cat.codes

# df2.head()

for column in columns_to_be_encoded:
    encoder=LabelEncoder()
    df2[column]=encoder.fit_transform(df2[column])

df2.head()
```

```python
#split the data
from sklearn.model_selection import train_test_split
X=df2.drop('Price (Euro)',axis=1)
Y=df2['Price (Euro)']

X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
#linear regression algorithm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#test the model and generate perfirmance metrics
y_pred=model.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)
```
